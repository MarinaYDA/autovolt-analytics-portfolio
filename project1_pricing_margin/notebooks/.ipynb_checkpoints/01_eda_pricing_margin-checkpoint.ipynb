```python
#Imports and settings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
```

```python
# Load data
data_path = "../data/project1_pricing_margin_clean.csv"
df = pd.read_csv(data_path)

df.head()
```

```python
#Quick overview
df.shape, df.columns
df.dtypes
df.isna().sum()
df.describe(include="all").T
```

```python
# Basic business checks
df["margin_pct"] = df["margin"] / df["total_price"] * 100

df[["total_price", "margin", "margin_pct"]].describe()
df.columns
```

```python
# Margin by vehicle_brand / vehicle_model
brand_summary = (
    df.groupby("vehicle_brand")
      .agg(
          jobs=("job_id", "count"),
          total_revenue=("total_price", "sum"),
          avg_margin=("margin", "mean"),
          avg_margin_pct=("margin_pct", "mean")
      )
      .sort_values("avg_margin", ascending=False)
)

brand_summary.head(20)
```

```python
model_summary = (
    df.groupby(["vehicle_brand", "vehicle_model"])
      .agg(
          jobs=("job_id", "count"),
          total_revenue=("total_price", "sum"),
          avg_margin=("margin", "mean"),
          avg_margin_pct=("margin_pct", "mean")
      )
      .query("jobs >= 10")  # ignore tiny samples
      .sort_values("avg_margin_pct", ascending=False)
)

model_summary.head(20)
```

```python
# Margin by region and job_type
region_job_summary = (
    df.groupby(["region", "job_type"])
      .agg(
          jobs=("job_id", "count"),
          total_revenue=("total_price", "sum"),
          avg_margin=("margin", "mean"),
          avg_margin_pct=("margin_pct", "mean")
      )
      .reset_index()
)

region_job_summary.head()
```

```python
# Simple charts
top_brands = brand_summary.head(8)

plt.figure()
top_brands["avg_margin"].plot(kind="bar")
plt.title("Average Margin by Brand (Top 8)")
plt.ylabel("Average margin ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
region_job_pivot = region_job_summary.pivot(index="region", columns="job_type", values="avg_margin_pct")
region_job_pivot.plot(kind="bar")
plt.title("Average Margin % by Region and Job Type")
plt.ylabel("Average margin %")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Repeat vs new customers
cust_summary = (
    df.groupby("is_repeat_customer")
      .agg(
          jobs=("job_id", "count"),
          total_revenue=("total_price", "sum"),
          avg_margin=("margin", "mean"),
          avg_margin_pct=("margin_pct", "mean")
      )
)

cust_summary
```
